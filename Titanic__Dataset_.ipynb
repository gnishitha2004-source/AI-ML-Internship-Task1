```python
# ===============================
# TASK 1: DATA CLEANING & PREPROCESSING
# Titanic Dataset - Ready-to-Run Template
# ===============================

# Step 1: Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from google.colab import files

# Step 2: Upload Titanic Dataset
print("ðŸ“‚ Please upload the Titanic dataset CSV (train.csv)")
uploaded = files.upload()

# Load dataset (replace with your file name if different)
df = pd.read_csv(list(uploaded.keys())[0])
print("âœ… Dataset loaded successfully!")
display(df.head())

# Step 3: Check dataset info
print("\nðŸ”Ž Dataset Info:")
df.info()
print("\nMissing Values Count:")
print(df.isnull().sum())

# Step 4: Handle Missing Values
print("\nðŸ›  Handling Missing Values...")

# Fill Age with median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Fill Embarked with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop Cabin (too many missing values)
df.drop(columns=['Cabin'], inplace=True)

print("âœ… Missing values handled!")
print(df.isnull().sum())

# Step 5: Encode Categorical Variables
print("\nðŸ”„ Encoding Categorical Variables...")
df = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True)
print("âœ… Encoding complete!")
display(df.head())

# Step 6: Scale Numerical Features
print("\nðŸ“Š Scaling Numerical Features...")
scaler = StandardScaler()
df[['Age','Fare']] = scaler.fit_transform(df[['Age','Fare']])
print("âœ… Scaling done!")
display(df[['Age','Fare']].head())

# Step 7: Outlier Detection (Boxplot)
print("\nðŸ“‰ Visualizing Outliers...")
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Fare'])
plt.title("Outliers in Fare")
plt.show()

# (Optional) Remove outliers using IQR method
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['Fare'] >= Q1 - 1.5*IQR) & (df['Fare'] <= Q3 + 1.5*IQR)]

print("âœ… Outliers handled! Shape after removal:", df.shape)

# Step 8: Save Cleaned Dataset
cleaned_file = "titanic_cleaned.csv"
df.to_csv(cleaned_file, index=False)
print(f"\nðŸ’¾ Cleaned dataset saved as {cleaned_file}")

# Download cleaned file
files.download(cleaned_file)

print("\nðŸŽ‰ Data Preprocessing Completed Successfully!")
```
